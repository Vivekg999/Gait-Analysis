# GAIT CYCLE

```python
from PIL import Image
import numpy as np
from skimage.io import imread, imshow
from skimage.io import imread_collection
import matplotlib.pyplot as plt
from scipy.signal import argrelextrema
import os
from skimage.transform import resize


def gait_cycle(images):
    
    # no of frames
    frames=0
    for t in images:
        frames+=1
    #print(frames)

    #no of pixels
    max=0
    l=[]
    for k in range(frames):
        count=0
        one=images[k]
        for i in  range(80,128):
            for j in range(88):
                if(one[i][j]):
                    count+=1
        l.extend([count])
    b=np.array(l)

    # printing graph
    a=range(frames)
    #plt.plot(a,l)
    plt.show()



    a=argrelextrema(b,np.greater)

    q=[0]
    for i in a:
        q.extend(b[i])
    q.extend([0])
    z=np.array(q)

    p=argrelextrema(z,np.greater)


    t=[]
    for i in a:
        t.extend(i)

    o=[]
    for j in p:
        o.extend(j)

    s=[]
    count=0
    for i in o:
        if(count>2):
            break
        s.append(t[i-1])
        count+=1
    diff=s[2]-s[0]   
    return diff,frames
```
